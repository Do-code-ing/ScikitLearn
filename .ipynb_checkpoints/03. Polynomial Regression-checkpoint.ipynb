# 3. Polynomial Regression

### 공부 시간에 따른 시험 점수 (우등생)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('PolynomialRegressionData.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## 3-1. 단순 선형 회귀 (Simple Linear Regression)

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X, y)
```

### 데이터 시각화 (전체)

```python
plt.scatter(X, y, color='blue')
plt.plot(X, reg.predict(X), color='green')
plt.title('Score by Hours (Genius)')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
reg.score(X, y)
```

## 3-2. 다항 회귀 (Polynomial Regression)

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4) # degree차 다항식
X_poly = poly_reg.fit_transform(X) # fit_transform() : 변환과 동시에 학습, 변환된 값은 X_poly 변수에 저장
X_poly[:5] # [x] -> [x^0, x^1, x^2]
```

```python
X[:5]
```

```python
poly_reg.get_feature_names_out() # 적용된 다항식 확인
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y) # 변환된 X 와 y 를 가지고 모델 생성 (학습)
```

### 데이터 시각화 (변환된 X 와 y)

```python
plt.scatter(X, y, color='blue')
# X 는 사실 그대로고, 다항식만 추가된 개념이기 때문에
# 새로 학습한 lin_reg 모델의 예측 값인 y 를 파라미터로 주되
# 앞서 다항식으로 변환이 있었으므로, 그 부분 인지시켜주기 위해
# predict 함수의 인수로 변환된 X_poly(=poly_reg.fit_transform(X)) 를 사용한다.
plt.plot(X, lin_reg.predict(poly_reg.fit_transform(X)), color='green')
plt.title('Score by Hours (Genius)')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
X_range = np.arange(min(X), max(X), 0.1) # X 의 최소값에서 최대값까지의 범위를 0.1 단위로 잘라서 데이터 생성
X_range
```

```python
X_range.shape
```

```python
X[:5]
```

```python
X_range = X_range.reshape(-1, 1) # reshape(row, col) row 가 -1인 경우 자동 계산, column 개수는 1개
X_range.shape
```

```python
X_range[:5]
```

```python
plt.scatter(X, y, color='blue')
# X 는 사실 그대로고, 다항식만 추가된 개념이기 때문에
# 새로 학습한 lin_reg 모델의 예측 값인 y 를 파라미터로 주되
# 앞서 다항식으로 변환이 있었으므로, 그 부분 인지시켜주기 위해 predict 함수의 인수로 변환된 X_poly 를 사용한다.
plt.plot(X_range, lin_reg.predict(poly_reg.fit_transform(X_range)), color='green')
plt.title('Score by Hours (Genius)')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

### 공부 시간에 따른 시험 성적 예측

```python
reg.predict([[2]]) # 2시간 공부했을 때 선형 회귀 모델의 예측
```

```python
lin_reg.predict(poly_reg.fit_transform([[2]])) # 2시간을 공부했을 때 다항 회귀 모델의 예측
```

```python
lin_reg.score(X_poly, y)
```
